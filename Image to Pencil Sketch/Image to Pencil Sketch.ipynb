### Importing libraries 

```python
import cv2
import matplotlib.pyplot as plt
```

### Reading the local image in the directory

```python
img = cv2.imread("test_pic.jpg")
```

### displaying BGR and converted BGR to RGB images

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Original image', size=18)
plt.imshow(img)
plt.axis("off")
plt.subplot(1,2,2)
plt.title('Crafted image', size=18)
plt.imshow(img[:,:,::-1])
plt.axis("off")
plt.show()
```

```python
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.axis(False)
plt.show()
```

### Convert to Grey Image

```python
grey_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(grey_img)
plt.axis(False)
plt.show()
```

### Invert Image

```python
invert_img=cv2.bitwise_not(grey_img)
plt.imshow(invert_img)
plt.axis(False)
plt.show()
```

### Blur image

```python
blur_img=cv2.GaussianBlur(invert_img, (111,111),0)
plt.imshow(blur_img)
plt.axis(False)
plt.show()
```

### invert Image again

```python
invblur_img=cv2.bitwise_not(blur_img)
plt.imshow(invblur_img)
plt.axis(False)
plt.show()
```

### The sketch can be obtained by performing bit-wise division

```python
sketch_img=cv2.divide(grey_img,invblur_img, scale=256.0)
plt.imshow(sketch_img)
plt.axis("off")
plt.show()
```

### Original Image and Sketch

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Original image', size=18)
plt.imshow(RGB_img)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Sketch', size=18)
rgb_sketch=cv2.cvtColor(sketch_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
